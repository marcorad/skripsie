```python
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib.widgets import Slider, Button
```

```python
N = 1024
n = np.array(range(0,N))

def square_wave(n, n_harmonics):
    N = (n.shape[0])
    x = np.zeros(N)
    for k in range(1, n_harmonics+1):
        x = x + np.sin(2 * np.pi * (2*k - 1) * 1/N * n) / (2*k - 1)
    return 2/np.pi * x + 0.5

def sawtooth_wave(n, n_harmonics):
    N = (n.shape[0])
    x = np.zeros(N)
    for k in range(1, n_harmonics+1):
        x = x + np.sin(2 * np.pi * k* 1/N * n) / k * (-1)**k
    return 0.5 - 1/np.pi * x

def triangle_wave(n, n_harmonics):
    N = (n.shape[0])
    x = np.zeros(N)
    for k in range(0, n_harmonics):
        x = x + np.sin(2 * np.pi * (2*k + 1) * 1/N * n) / (2*k+1)**2 * (-1)**k
    return 8/(np.pi**2) * x

```

```python
plt.plot(n, square_wave(n, 16))
plt.show()
```
